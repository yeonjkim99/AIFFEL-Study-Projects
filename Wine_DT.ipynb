```python
# (1) 필요한 모듈 import
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

# Model에 따라 이부분 변경======================================
from sklearn.tree import DecisionTreeClassifier
decision_tree = DecisionTreeClassifier
print('00 estimator_type ======================================')
print(decision_tree._estimator_type)
print()

# (2) 데이터 준비
wine = load_wine()
print('01 variable, method 및 key 확인 ========================')
print(dir(wine))
print(wine.keys())
print()

# (3) 데이터 이해하기
wine_data = wine.data
print('02 Data 구성 및 첫번째 data 확인 =======================')
print(wine_data.shape)
print(wine_data[0])
print()

print('03 Data의 future 이름 확인==============================')
print(wine.feature_names)
print()

wine_label = wine.target
print('04 Data의 label 확인 ===================================')
print(wine_label.shape)
print(wine_label)
print()

print('05 Data의 target name 확인==============================')
print(wine.target_names)
print()

print('06 Data의 description 확인==============================')
print(wine.DESCR)
print()


# (4) train, test 데이터 분리
X_train, X_test, y_train, y_test = train_test_split(wine_data, 
                                                    wine_label, 
                                                    test_size=0.2, 
                                                    random_state=21)  # 숫자변경====
print('07 train data 및 test data 정보 확인 ===================')
print('X_train 개수: ', len(X_train),', X_test 개수: ', len(X_test))
print([X_train.shape, y_train.shape])
print([X_test.shape, y_test.shape])
print(y_train)
print(y_test)
print()


# (5) 모델 학습 및 학습결과 확인 
# Model에 따라 이부분 변경======================================
decision_tree = DecisionTreeClassifier(random_state=32)
decision_tree.fit(X_train, y_train)
y_pred = decision_tree.predict(X_test)

print('08 test 결과 확인 ======================================')
print(y_pred)
print()

# (6) 성능 평가
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

print('09 학습결과 accuracy 및 confusion matrix 확인 ==========')
accuracy = accuracy_score(y_test, y_pred)
print(accuracy)
print(confusion_matrix(y_test, y_pred))  # 정답, 오답 갯수 확인, 성능지표값은 별도 계산필요
print()

print('10 학습결과 보고서 =====================================')
print(classification_report(y_test, y_pred))
```
